```python
#importando bibliotecas
import pandas as pd
import plotly.express as px
import plotly.io as pio
from plotly.offline import plot, iplot, init_notebook_mode
import plotly.graph_objs as go
init_notebook_mode(connected=True)
pio.renderers.default = "svg"
```

# Importando a base de dados 

```python
tabela = pd.read_csv("ClientesBanco.csv", encoding="latin1")
tabela = tabela.drop("CLIENTNUM", axis=1)  #coluna inútil
display(tabela)
```

# Agora vamos tratar valores vazios e exibir um resumo das colunas da base de dados

```python
tabela = tabela.dropna()
display(tabela.info())

display(tabela.describe().round(1))
```

# Vamos avaliar como está a divisão entre Clientes x Cancelados

```python
qtde_categoria = tabela["Categoria"].value_counts()
display(qtde_categoria)

fig = px.pie(qtde_categoria,
             values='Categoria',
             names=['Cliente', 'Cancelado'])
fig.show()
```

# Descobrindo o motivo de cancelamento
  - Podemos olhar a comparação entre Clientes e Cancelados em cada uma das colunas da nossa base de dados, para ver se essa informação traz algum insight

```python
for coluna in tabela:
    grafico = px.histogram(
        tabela,
        x=coluna,
        color="Categoria",
        width=500,
        height=500,
    )
    grafico.show()
```

# Informações retiradas da análise

- Quanto mais produtos contratados um cliente tem, menor a chance dele cancelar
- Quanto mais transações e quanto maior o valor de transação, menor a chance dele cancelar
- Quanto maior a quantidade de contatos que a pessoa teve que fazer, maior a chance dela cancelar
